```python
import re

words = re.split('[^A-Za-z]+', raw.lower()) # Remove empty strings

# Print length of list
print("Total number of words in the corpus: ",len(words))

# Create set of all unique words, this throws away any information about frequency however
gram1 = set(words)

print("Number of unigrams:",len(gram1))

# Instead of printing all the elements in the set, create an iterator and print 20 elements only
word_pairs = [(words[i], words[i+1]) for i in range(len(words)-1)]

gram2 = set(word_pairs)
print("Number of possible bigrams:",len(gram1)**2)
print("Number of bigrams:",len(gram2))

word_triplets = [(words[i], words[i+1], words[i+2]) for i in range(len(words)-2)]

gram3= set(word_triplets)

print("Number of trigrams:",len(gram3))

word_quads = [(words[i], words[i+1], words[i+2], words[i+3]) for i in range(len(words)-3)]

gram4= set(word_quads)

print("Number of quadgrams:",len(gram4))
```

```python
#MLE for Unigram
print("MLE for Unigram: \n\n")

tokens = len(words)

for word in words:
    if word not in Frequency.keys():
        Frequency[word] = words.count(word)
#print(wordFreq)
for i,j in reversed(sorted(Frequency.items(), key = itemgetter(1))):
    print('P( ' + i +' ) = ' + str(j/tokens) )
```

```python
#MlE for Bigram
print("MLE for Bigram:")
print()
gram2 = {}
for sentence in train_data:
    wrdList = ngrams(word_tokenize(sentence),2)
    
    for bigram in wrdList:
        if bigram not in gram2:
            gram2[bigram] = 1
        else:
            v = gram2[bigram]
            gram2[bigram] = v + 1
for k,v in reversed(sorted(gram2.items(), key = itemgetter(1))):
    print('P( ' +k[1] + ' / '+ k[0] +' ) = ' + str(v/Frequency[k[0]]))
```
